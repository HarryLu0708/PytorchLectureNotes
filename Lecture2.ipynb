## Simple Linear Model(x*w)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = [1.0,2.0,3.0]
y_data = [2.0,4.0,6.0]
```

```python
def forward(x,w):
  return x*w
```

```python
def loss(y_pred,y):
  return (y_pred-y)*(y_pred-y)
```

```python
w_list = []
mse_list = []
```

```python
for w in np.arange(0.0,4.1,0.1):
  print("w=",w)
  l_sum = 0
  for x,y in zip(x_data,y_data):
    y_pred = forward(x,w)
    l_val = loss(y_pred,y)
    l_sum += l_val
    print(x,y,y_pred,l_val)
  print('MSE:',l_sum/3)
  w_list.append(w)
  mse_list.append(l_sum/3)
```

```python
plt.plot(w_list,mse_list)
plt.ylabel('Loss')
plt.xlabel('w')
plt.show()
```

## Assignment: Linear Model(x*w+b)

```python
def linear_forward(x,w,b):
  return x*w+b
```

```python
def linear_loss(y_pred,y):
  return (y_pred-y)**2
```

```python
w_list = np.arange(0.0,4.1,0.1)
mse_matrix = np.zeros((len(np.arange(0.0,4.1,0.1)),len(np.arange(0.0,4.1,0.1))))
b_list = np.arange(-2.0,2.1,0.1)
```

```python
for i,w in enumerate(np.arange(0.0,4.1,0.1)):
  for j,b in enumerate(np.arange(-2.0,2.0,0.1)):
    l_sum = 0
    for x,y in zip(x_data,y_data):
      y_pred = linear_forward(x,w,b)
      l_val = linear_loss(y_pred,y)
      l_sum += l_val
      print(x,y,y_pred,l_val)
    print('Loss:',l_sum/3)
    mse_matrix[i,j] = l_sum/3
```

```python
W, B = np.meshgrid(w_list, b_list)
fig = plt.figure('Linear Model Cost Value')
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W, B, mse_matrix.T, cmap='viridis')
ax.set_xlabel('w')
ax.set_ylabel('b')
ax.set_zlabel('loss')
plt.show()
```
