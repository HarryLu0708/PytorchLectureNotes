## Multiple-Dimensions Input

```python
import torch
```

```python
class MDModel(torch.nn.Module):
  def __init__(self):
    super(MDModel,self).__init__()
    self.linear1 = torch.nn.Linear(8,6)
    self.linear2 = torch.nn.Linear(6,4)
    self.linear3 = torch.nn.Linear(4,2)
    self.linear4 = torch.nn.Linear(2,1)
    self.activation = torch.nn.Sigmoid()

  def forward(self,x):
    y_pred = self.activation(self.linear1(x))
    y_pred = self.activation(self.linear2(y_pred))
    y_pred = self.activation(self.linear3(y_pred))
    y_pred = self.activation(self.linear4(y_pred))
    return y_pred
```

```python
model = MDModel()
```

```python
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)
criterion = torch.nn.BCELoss()
```

```python
import numpy as np

xy = np.loadtxt("/content/diabetes.csv.gz",delimiter=",",dtype=np.float32)
x_data = torch.from_numpy(xy[:,:-1])
y_data = torch.from_numpy(xy[:,[-1]])
```

```python
for epoch in range(1000):
  y_pred = model(x_data)
  loss = criterion(y_pred,y_data)
  print(epoch,loss.item())
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```
