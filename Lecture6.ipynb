## Logistics Regression

```python
import torch
```

```python
x_data = [[1.0],[2.0],[3.0]]
y_data = [[0.0],[0.0],[1.0]]
```

```python
x_data = torch.as_tensor(x_data)
y_data = torch.as_tensor(y_data)
```

```python
x_data
```

```python
class LogisticsModel(torch.nn.Module):
  def __init__(self):
    super(LogisticsModel,self).__init__()
    self.linear = torch.nn.Linear(1,1)

  def forward(self,x):
    y_pred = torch.sigmoid(self.linear(x))
    return y_pred
```

```python
model = LogisticsModel()
```

```python
criterion = torch.nn.BCELoss(size_average=False)
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)
```

```python
# start training!
for epoch in range(2000):
  y_pred = model(x_data)
  loss = criterion(y_pred,y_data)
  print(epoch,loss.item())
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```

```python
x_test = torch.Tensor([5.0])
y_test = model(x_test)
print(y_test.data)
```
