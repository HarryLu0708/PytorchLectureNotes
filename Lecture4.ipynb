## Back Propagation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import torch
```

```python
x_data = [1.0,2.0,3.0]
y_data = [2.0,4.0,6.0]
w = torch.Tensor([1.0])
w.requires_grad = True
```

```python
def forward(x):
  return x*w
```

```python
def loss(x,y):
  y_pred = forward(x)
  return (y_pred-y)**2
```

```python
print(4,forward(4).item())
```

```python
for epoch in range(100):
  for x,y in zip(x_data,y_data):
    l = loss(x,y)
    l.backward() # do backward
    print(x,y,w.grad.item())
    w.data = w.data-0.01*w.grad.data # update weights
    w.grad.data.zero_() # clean the gradient
  print("progress:",epoch,l.item())
print(4,forward(4).item())
```
