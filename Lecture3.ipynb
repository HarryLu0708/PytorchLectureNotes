## Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = [1.0,2.0,3.0]
y_data = [2.0,4.0,6.0]
w = 1.0
```

```python
def forward(x):
  return x*w
```

```python
def cost(xs,ys):
  cost = 0
  for x,y in zip(xs,ys):
    y_pred = forward(x)
    cost += (y_pred-y)**2
  return cost/len(xs)
```

```python
def gradient(xs,ys):
  grad = 0
  for x,y in zip(xs,ys):
    grad += 2*x*(x*w-y)
  return grad/len(xs)
```

```python
print(4,forward(4))
```

```python
for epoch in range(100):
  cost_val = cost(x_data,y_data)
  grad = gradient(x_data,y_data)
  w -= 0.01*grad
  print('Epoch:',epoch,'w=',w,'loss=',cost_val)
print(4,forward(4))
```

## SGD

```python
def loss(x,y):
  y_pred = forward(x)
  return (y_pred-y)**2
```

```python
def gradient(x,y):
  return 2*x*(x*w-y)
```

```python
w = 1.0
```

```python
print(4,forward(4))
```

```python
for epoch in range(100):
  for x,y in zip(x_data,y_data):
    grad = gradient(x,y)
    w -= grad*0.01
    print(x,y,grad)
    l = loss(x,y)
  print("Epoch:",epoch,"w:",w,"loss",l)
```
