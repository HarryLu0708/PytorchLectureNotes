## Datasets and DataLoaders

```python
import torch
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
```

```python
import numpy as np
```

```python
class DiabetesDataset(Dataset):
  def __init__(self):
    xy = np.loadtxt("/content/diabetes.csv.gz",delimiter=",",dtype=np.float32)
    self.len = xy.shape[0]
    self.x_data = torch.from_numpy(xy[:,:-1])
    self.y_data = torch.from_numpy(xy[:,[-1]])

  def __getitem__(self,index):
    return self.x_data[index], self.y_data[index]

  def __len__(self):
    return self.len
```

```python
dataset = DiabetesDataset()
train_loader = DataLoader(dataset=dataset,batch_size=32,shuffle=True,num_workers=2)
```

```python
# POKEMON!
class LogisticsModel(torch.nn.Module):
  def __init__(self):
    super(LogisticsModel,self).__init__()
    self.activation = torch.nn.Sigmoid()
    self.linear1 = torch.nn.Linear(8,6)
    self.linear2 = torch.nn.Linear(6,4)
    self.linear3 = torch.nn.Linear(4,2)
    self.linear4 = torch.nn.Linear(2,1)

  def forward(self,x):
    y_pred = self.activation(self.linear1(x))
    y_pred = self.activation(self.linear2(y_pred))
    y_pred = self.activation(self.linear3(y_pred))
    y_pred = self.activation(self.linear4(y_pred))
    return y_pred
```

```python
model = LogisticsModel()
optimizer = torch.optim.Adam(model.parameters(),lr=0.01)
criterion = torch.nn.BCELoss()
```

```python
# Start training!
for epoch in range(1000):
  for i,data in enumerate(train_loader, 0):
    inputs,labels = data
    y_pred = model(inputs)
    loss = criterion(y_pred,labels)
    print(epoch,i,loss.item())
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```
